Для варіанту, заданого викладачем, задати і візуалізувати граф за
допомогою коду, наведеного у цій роботі.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_nodes_from([1, 2, 3, 4, 5, 6])
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 3), (2, 5), (3, 5), (4, 5), (4, 6)])

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=12, font_weight='bold', edge_color='black', width=1.5, arrowsize=20)

plt.title("Ациклічний граф")
plt.show()
```
